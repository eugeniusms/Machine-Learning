```python
%matplotlib widget

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

# Generate Data

```python
jumlah_data = 100

y = np.array([i*0.1+np.random.randn() for i in range(jumlah_data)])
x = np.array([i*0.1 for i in range(jumlah_data)])
```

### 2. Fungsi Linear

```python
def ucup_linear(x, gradien):
    y = gradien*x
    return y
```

Menghitung persamaan linear awal

```python
x_prediction = np.array([0, 10])
m_init = 5
y_prediction = ucup_linear(x_prediction, m_init)
```

```python
# plt.scatter(x, y)
# plt.plot(x_prediction, y_prediction, "Red")
# plt.axis([0,10,0,10])
# plt.show()
```

<img src="images/detecting_gradient_error.jpg" width="600" height="300">

```python
m_prediction = m_init
m_list_prediction = []
y_list_prediction = []
x_list_prediction = []
learning_rate = 0.1

for i in range(1, jumlah_data):
    y_prediction = ucup_linear(x[i], m_prediction)
    y_actual = y[i]

    error = y_actual - y_prediction
    delta_m = learning_rate * error/x[i]

    m_prediction = m_prediction + delta_m

    m_list_prediction.append(m_prediction)
    y_list_prediction.append(ucup_linear(np.array([0, 10]), m_prediction))
    x_list_prediction.append(np.array([0, 10]))
```

Visualisasi Prediksi

```python
fig = plt.figure(figsize=(4,4))
line, = plt.plot([], [], "Red")

def animate(frame_num):
    x = x_list_prediction[frame_num]
    y = y_list_prediction[frame_num]
    line.set_data(x, y)
    return line

plt.scatter(x, y)
plt.axis([0, 10, 0, 10])
anim = FuncAnimation(fig, animate, frames=100, interval=100, repeat=False)
plt.show()
```
